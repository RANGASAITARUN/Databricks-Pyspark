```
import pyspark
from pyspark.sql import SparkSession 
from pyspark.sql.functions import *
```

```

data = [('James','','Smith','1991-04-01','M',3000),
  ('Michael','Rose','','2000-05-19','M',4000),
  ('Robert','','Williams','1978-09-05','M',4000),
  ('Maria','Anne','Jones','1967-12-01','F',4000),
  ('Jen','Mary','Brown','1980-02-17','F',-1)
]

columns = ["firstname","middlename","lastname","dob","gender","salary"]
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()
df = spark.createDataFrame(data=data, schema = columns)
```

```
display(df)
```

Select columns

```
df1=df.select("middlename")
display(df1)
```

```
df2=df.select(["middlename","lastname"])
display(df2)
```

withCoulmnRenamed

```
df3=df.withColumnRenamed("middlename","Middlename")
display(df3)
```

```
df4=df.orderBy(df.salary.asc())
display(df4)
```

```
df5=df.orderBy(df.salary.desc())
display(df5)
```

Groupby

```
df6=df.groupBy("gender").sum("salary")
display(df6)
```

Filter

```
df7=df.filter(df.salary >=3000)  >=,<=,==,&,||,!
display(df7)
```

Substring

```
df8=df.withColumn("Year",substring(df.dob,0,4))
display(df8)
```

Drop

```
df9=df.drop("salary")
display(df9)
```

dropduplicates

```
df10=df.dropDuplicates()
display(df10)
```

Collect

```
for i in df.collect():
  print(i)
```

```
for i in df.collect()[0]:
  print(i)
```

```
for i in df.collect():
  print(i[0])
```

regexp_replace/transalate

```
df11=df.withColumn("dob",regexp_replace("dob","-","/"))
display(df11)
```

```
# if table conatins null value means it replace with 0
display(df.na.fill(value=0))
# display(df12)
```

whenotherwise

```
df12=df.withColumn("gender",when(df.gender=="M","Male")
                               .when(df.gender=="F","Female")
                                .otherwise(df.gender))
display(df12)
```

Joinsinner,left,right,outer

```
emp = [(1,"Smith",-1,"2018","10","M",3000), \
    (2,"Rose",1,"2010","20","M",4000), \
    (3,"Williams",1,"2010","10","M",1000), \
    (4,"Jones",2,"2005","10","F",2000), \
    (5,"Brown",2,"2010","40","",-1), \
      (6,"Brown",2,"2010","50","",-1) \
  ]
empColumns = ["emp_id","name","superior_emp_id","year_joined", \
       "emp_dept_id","gender","salary"]

empDF = spark.createDataFrame(data=emp, schema = empColumns)
empDF.printSchema()
empDF.show(truncate=False)

dept = [("Finance",10), \
    ("Marketing",20), \
    ("Sales",30), \
    ("IT",40) \
  ]
deptColumns = ["dept_name","dept_id"]
deptDF = spark.createDataFrame(data=dept, schema = deptColumns)
deptDF.printSchema()
deptDF.show(truncate=False)
```

```
display(empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"inner"))
```

```
display(empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"left"))
```

```
display(empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"right"))
```

```
display(empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"outer"))
```

```
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()

simpleData = [("James","Sales","NY",90000,34,10000), \
    ("Michael","Sales","NY",86000,56,20000), \
    ("Robert","Sales","CA",81000,30,23000), \
    ("Maria","Finance","CA",90000,24,23000) \
  ]

columns= ["employee_name","department","state","salary","age","bonus"]
df14 = spark.createDataFrame(data = simpleData, schema = columns)
df14.printSchema()
df14.show(truncate=False)


simpleData2 = [("James","Sales","NY",90000,34,10000), \
    ("Maria","Finance","CA",90000,24,23000), \
    ("Jen","Finance","NY",79000,53,15000), \
    ("Jeff","Marketing","CA",80000,25,18000), \
    ("Kumar","Marketing","NY",91000,50,21000) \
  ]
columns2= ["employee_name","department","state","salary","age","bonus"]

df15= spark.createDataFrame(data = simpleData2, schema = columns2)

df15.printSchema()
df15.show(truncate=False)
```

Window partition

```
spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()

simpleData = (("James", "Sales", 3000), \
    ("Michael", "Sales", 4600),  \
    ("Robert", "Sales", 4100),   \
    ("Maria", "Finance", 3000),  \
    ("James", "Sales", 3000),    \
    ("Scott", "Finance", 3300),  \
    ("Jen", "Finance", 3900),    \
    ("Jeff", "Marketing", 3000), \
    ("Kumar", "Marketing", 2000),\
    ("Saif", "Sales", 4100) \
  )
 
columns= ["employee_name", "department", "salary"]
df16= spark.createDataFrame(data = simpleData, schema = columns)
df16.printSchema()
df16.show(truncate=False)
```

```

from pyspark.sql.window import Window
from pyspark.sql.functions import row_number
windowSpec  = Window.partitionBy("department").orderBy("salary")

display(df16.withColumn("row_number",row_number().over(windowSpec)))
```

```
display(df16.withColumn("rank",rank().over(windowSpec)))
```

```
display(df16.withColumn("dense_rank",dense_rank().over(windowSpec)))
```
